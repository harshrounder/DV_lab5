```python
import pandas as pd

# Load the Adult dataset into a Pandas dataframe

column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 
                'marital-status', 'occupation', 'relationship', 'race', 'sex', 
                'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header=None)


df.columns = column_names




```

Creating A scatter plot

```python

# Subset the dataframe to only numeric columns
numeric_cols = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
numeric_df = df[numeric_cols]

# Add a column to the numeric dataframe indicating the income level
numeric_df['income'] = df['income']

# Create a pair plot of the numeric dataframe with different colors for different income levels
sns.pairplot(data=numeric_df, hue='income', palette='bright')
plt.show()
```

From the scatter plot we see that both the class are not linearly sepearable also in some cases one class is outliar for the whole dataset

```python
import random
categorical_cols = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country', 'income']
for col in categorical_cols:
    plt.figure(figsize=(15,5))
    colors = ["red","yellow","blue","purple","orange","black"]
    df[col].value_counts().plot(kind='bar',color =random.choice(colors))
    plt.title(col)
    plt.show()
```

These bar graphs show the data distribution for the numerical columns, so from here we can deduce many inferences like in sex graph we see that males are more than females and so on.

Correlation matrix for numeric columns


```python

numeric_cols = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

bar chart of income by education level

```python

plt.figure(figsize=(15,5))
sns.countplot(x='education', hue='income', data=df)
plt.title('Income by Education Level')
plt.show()
```

box plot of age by income

```python

plt.figure(figsize=(10,5))
sns.boxplot(x='income', y='age', data=df)
plt.title('Age by Income')
plt.show()
```

Creating a stacked bar chart of marital status by income and sex

```python

plt.figure(figsize=(10,5))
df.groupby(['marital-status', 'sex', 'income']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Marital Status by Income and Sex')
plt.show()
```
